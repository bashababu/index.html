```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(1,6)
y=x**2 # here y contains Squares of x ndarray
z=x**3 # here z contains Cubes of x ndarray
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
# Apply legend()
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(1,6)
k=np.array([2,6,9,5,3])
y=x**2 # here y contains Squares of x ndarray
z=x**3 # here z contains Cubes of x ndarray
plt.plot(x,k,label="LINEAR")
plt.plot(x,y,label="SQUARES")
plt.plot(x,z,label="CUBES")
plt.legend()
plt.show()
```

```python
# Apply legend() with with label attribute
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(1,6)
k=np.array([2,6,9,5,3])
y=x**2 # here y contains Squares of x ndarray
z=x**3 # here z contains Cubes of x ndarray
lab=["Linear","Squares","Cubes"]
plt.plot(x,k)
plt.plot(x,y)
plt.plot(x,z)
plt.legend(labels=lab,loc="upper center",edgecolor="r",fontsize=20)
plt.show()
```
